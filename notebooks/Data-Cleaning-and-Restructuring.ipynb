# Data Cleaning & Restructuring

```python
dallas_city = pd.read_csv('../dataset/raw/Dallas_City_Data_2021.csv', index_col='ID')
dallas_city.head(5)
```

### Make Column names more self-explanatory

```python
dallas_city.rename(columns={'Number': 'Street_Number', 
                            'Street':'Street_Name', 
                            'Side': 'Street_Side'}, inplace=True)
dallas_city.head()
```

### Confirm single-valued columns

```python
dallas_city[['City', 'State', 'Country', 'Timezone']]
```

### Note: 

As shown in the previous jupyter notebook in basic exploration of the US-Accidents data and confirmed in the above cell, some columns share the same values all throughout this subset since we extracted Dallas City dataset. <hr>

### Dropping columns

```python
dallas_city.drop(columns=['City', 'State', 'Country', 'Timezone' ], inplace=True)
dallas_city.head(5)
```

```python
print(f'The number of columns is reduced to {dallas_city.shape[1]}.')
```

### Checking Missing Values in Dallas City subset 

```python
na_count = dallas_city.isna().sum()
missing_val = na_count[dallas_city.isna().sum()>0]
missing_val
```

```python
missing_val = missing_val.reset_index()
missing_val
```

```python
missing_val.rename(columns={'index':'Field_NaN', 0:'Count'}, inplace=True)
missing_val
```

## Investigating Columns

### 1. Start_Time & End_Time

Confirm if _'Start_Time'_ and _'End_Time'_ are of valid Datetime type

```python
# If this code returns error then the Start_Time is not a valid datetime 
dallas_city.loc['A-11786', 'Start_Time'].day_name()
```

### Convert to Datetime type

```python
dallas_city['Start_Time'] = pd.to_datetime(dallas_city['Start_Time'])
dallas_city['End_Time'] = pd.to_datetime(dallas_city['End_Time'])

dallas_city.head(5)
```

```python
df1 = (dallas_city['End_Time'] - dallas_city['Start_Time'])
```

### 2. Description: Confirm whether there is a uniform parttern we can extract

```python
dallas_city.loc['A-2844612':'A-2845068', 'Description']
```

### Note: Looking at a subset of the 'Description' series, we can infer that it might follow a certain pattern that may or may not include road blockage and traffic condition details.

Hence, we will look at this more closely by splitting the entire string object and expand into new columns.

```python
df = dallas_city['Description'].str.split('.', 1, expand=True)
df
```

```python
df.rename(columns={1:'Impact'}, inplace=True)
```

```python
df[['Description', 'Status']] = df[0].str.rsplit(' - ', 1, expand=True)
df.drop(columns=[0], inplace=True)
```

```python
pd.DataFrame(df['Status'].value_counts()).reset_index().rename(columns={'index':'Status', 'Status':'Count'})
```

```python
pd.DataFrame(df['Impact'].value_counts()).reset_index().rename(columns={'index':'Impact', 'Impact':'Count'})
```

```python
df['Description'].value_counts()
```

```python
df.isna().sum()
```

### Note: Although there are important details in the description pertaining to traffic impact, road blockage and type of incident , it is really hard to tell and extract these as it does not follow a particular pattern in contrary to what we assumed.

### 2. Street Number

```python
dallas_city['Street_Number'].describe()
```

### Note: The Street number is considered as a **numerical data** rather than categorical.

This is because the statistical summary such as Mean, Median and Standard Deviation can be calculated in this series, which should not be the case since we do not expect to make calculations on Street Numbers.

### Replacing NAN values in the Street Number Attribute

```python
dallas_city['Street_Number'].replace(np.nan, "Not Available", inplace=True) 
```

```python
dallas_city['Street_Number'].describe()
```

By changing the np.nan value, which is of type float64, we also changed the entire series as a categorical data. We expect to not see the _'Street_Number'_ column when displaying the statistical summary of the numerical data.

```python
dallas_city.describe()
```
