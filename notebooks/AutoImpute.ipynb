```python
acclist = []
for iteration in range(kernel.iteration_count()+1):
    wc_na_count = kernel.na_counts[17]
    compdat = kernel.complete_data(dataset=0,iteration=iteration)
    
    # Record the accuracy of the imputations of species.
    acclist.append(
      round(1-sum(compdat['Wind_Chill(F)'] != sample['Wind_Chill(F)'])/wc_na_count,2)
    )

# acclist shows the accuracy of the imputations
# over the iterations.
print(acclist)
```
