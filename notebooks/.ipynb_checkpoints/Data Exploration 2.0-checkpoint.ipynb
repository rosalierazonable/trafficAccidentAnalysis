# Data Exploration for Dallas City and County Data Subsets

This is to explore the **594 data discrepancy** between the two subsets and to understand the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
```

```python
dallas_city = pd.read_csv('../dataset/Dallas_City_Data_2021.csv')
dallas_city.head(5)
```

```python
dallas_city.columns
```

```python
# Check statistics 
dallas_city.describe()
```

```python
dallas_city.hist(bins=60, figsize=(20,10), color='green')
```

```python
style.use('ggplot')
plt.hist(dallas_city.Severity, bins=[1,2,3,4,5], edgecolor='black', rwidth=0.8, color='blue')
plt.title("Frequency vs Severity")
plt.xlabel("Severity")
plt.ylabel("Frequency of Accidents")
```

```python
plt.style.use('_mpl-gallery')
fig, ax = plt.subplots()
ax.plot(dallas_city.Start_Time,dallas_city.Severity)

plt.show()
```
