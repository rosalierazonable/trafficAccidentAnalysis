## Loading Dataset to a dataframe variable

```python
us_data = pd.read_csv("../dataset/US_Accidents_Dec21_updated.csv")
us_data.head(5)
```

```python
# Confirm available attributes or columns
us_data.columns
```

```python
type(us_data)
```

```python
pd.Series(us_data["State"]).value_counts().head(3)
```

## Determine how many Traffic Accident records are available in the Texas State

```python
texas_data = us_data[us_data["State"]=="TX"]
texas_data.shape
```

```python
texas_data.head(5)
```

## Convert Dataframe object to CSV file
<hr>

```python
texas_data.to_csv('../dataset/Texas_Accidents_Data_2021.csv')
```

## Checking Traffic Incidents in Dallas City

```python
pd.Series(texas_data["City"]).value_counts()
```

## Checking Traffic Incidents in Dallas County

```python
pd.Series(texas_data["County"]).value_counts()
```


## Explore Dalla City Traffic Accidents data
<hr>

```python
dallas_city_data = texas_data[texas_data["City"]=="Dallas"]
dallas_city_data.shape # Confirm the available number of traffic accident records of Dallas City
```

```python
dallas_city_data.hist(bins=60, figsize=(60,30))
```


## Explore the available data of Dallas County
<hr>

```python
dallas_county_data = texas_data[texas_data["County"]=="Dallas"]
dallas_county_data.shape # Confirm the available number of traffic accident records of Dallas County
```

```python
dallas_county_data.hist(bins=60, figsize=(60,30))
```


## Check accident reports breakdown by City
<hr>

```python
pd.Series(dallas_county_data["City"]).value_counts()
```

**Notice** that here there are only **41,021 Traffic Accident** records for Dallas City based on the **County** dataframe, but the **Dallas City** dataframe has **41,615 records** with **594 discrepancy** between these two subsets.<br>
To understand, we will look at these individual data subsets more closely.
