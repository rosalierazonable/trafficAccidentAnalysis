# Data Cleaning & Restructuring for Dallas City and County Data Subsets

This is to explore the **594 data discrepancy** between the two subsets and to understand the data.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
```

```python
# To Display all columns 
pd.set_option('display.max_columns', None)

style.use(['science', 'notebook', 'grid'])
```

```python
dallas_city = pd.read_csv('../dataset/raw/Dallas_City_Data_2021.csv', index_col='ID')
dallas_city.head(5)
```

```python
# Closer look at the names of the columns or attributes
dallas_city.columns
```

### Make Columns more self-explanatory

```python
dallas_city.rename(columns={'Number': 'Street_Number', 
                            'Street':'Street_Name', 
                            'Side': 'Street_Side'}, inplace=True)
dallas_city.head()
```

```python
dallas_city[['City', 'State', 'Country', 'Timezone']]
```

### Dropping columns:

Note: As shown in the previous jupyter notebook in basic exploration of the US-Accidents data and confirmed in the above cell, some columns share the same values all throughout this subset since we extracted Dallas City dataset.

```python
dallas_city.drop(columns=['City', 'State', 'Country', 'Timezone' ], inplace=True)
dallas_city
```

```python
dallas_city.shape
```

### Checking Missing Values in Dallas City subset <hr>

```python
dallas_city.isna().sum()
```

```python
# Replacing NAN values in the Street Number Attribute
dallas_city.Number.replace(np.nan, "Not Available", inplace=True) 
```

```python
# Check if there are duplicate entries or records
dallas_city[dallas_city.duplicated()]
```
