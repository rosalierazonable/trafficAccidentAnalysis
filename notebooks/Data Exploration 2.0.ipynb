# Data Exploration for Dallas City and County Data Subsets

This is to explore the **594 data discrepancy** between the two subsets and to understand the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
```

```python
# To Display all 47 columns 
pd.set_option('display.max_columns', None)

style.use(['science', 'notebook', 'grid'])
```

```python
dallas_city = pd.read_csv('../dataset/raw/Dallas_City_Data_2021.csv')
dallas_city.head(5)
```

```python
# Closer look at the names of the columns or attributes
dallas_city.columns
```

## Examine why there is a discrepancy in traffic accident records between Dallas City and County Dataframes 
<hr>

```python
discrepancy = dallas_city[dallas_city['County']!="Dallas"]
discrepancy.head(5)
```

```python
pd.Series(discrepancy.County).value_counts()
```

### By looking at the Dallas City records whose County value is not Dallas, we realized that the 594 is not a discrepancy at all. 

These **594 records** are in fact Dallas City traffic accident records that occured in different Counties: 
1. Collin (383), 
2. Tarrant (130) and 
3. Denton (81)

## Using Advanced Matplotlib API <hr>

```python
fig, ax = plt.subplots(1, 1, figsize=(15,4))
ax.hist(dallas_city.Severity, edgecolor='black', bins=10)
ax.set_xlabel('Accident Severity')
ax.set_ylabel('Frequency of Accidents')
ax.set_title('Frequency vs Severity')
plt.show()
```

```python
dallas_city.hist(bins=30, figsize=(30,15), color='green', edgecolor='black')
plt.show()
```

```python
plt.hist(dallas_city.Severity, bins=[1,2,3,4,5], edgecolor='black', rwidth=0.8, color='blue')
plt.title("Frequency vs Severity")
plt.xlabel("Severity")
plt.ylabel("Frequency of Accidents")
plt.show()
```
